```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os,json

FOLDER = "Data/"
os.makedirs(FOLDER, exist_ok=True)
os.listdir(FOLDER)
```

```python
df1=pd.read_csv('Data/superhero_info.csv')
df2=pd.read_csv('Data/superhero_powers.csv')
```

```python
df1['Hero|Publisher'].str.split('|',expand=True)
```

```python
df1[['hero','publisher']]= df1['Hero|Publisher'].str.split('|',expand=True)
```

```python
df1.head(2)
```

```python
df1=df1.drop(columns=['Hero|Publisher'])
df1.head(2)
```

```python

to_replace = ['{','}']

for char in to_replace:
    df1['Measurements'] = df1['Measurements'].str.replace(char,'',regex=False)
    
df1['Measurements'].head()

```

```python
df1[['h1','w1']]= df1['Measurements'].str.split(',',expand=True)
df1=df1.drop(columns=['Measurements'])
df1.head(2)
```

```python
df1[['whaat','height(cm)','YAA']]= df1['h1'].str.split(' ',expand=True)
df1=df1.drop(['h1','YAA','whaat'],axis=1)
df1.head(2)
```

```python
df1[['w','woah','weight(kg)','w2']]= df1['w1'].str.split(' ',expand=True)
df1=df1.drop(['w','woah','w2','w1'],axis=1)
df1.head(2)
```

```python

to_replace = ["'"]

for char in to_replace:
    df1['height(cm)'] = df1['height(cm)'].str.replace(char,'',regex=False)
    df1['weight(kg)'] = df1['weight(kg)'].str.replace(char,'',regex=False)
    
df1.head()

```
