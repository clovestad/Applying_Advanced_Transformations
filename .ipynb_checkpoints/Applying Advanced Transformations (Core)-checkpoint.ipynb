```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
import os,json

FOLDER = "Data/"
os.makedirs(FOLDER, exist_ok=True)
os.listdir(FOLDER)
```

```python
df1=pd.read_csv('Data/superhero_info.csv')
df2=pd.read_csv('Data/superhero_powers.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1['Hero|Publisher'].str.split('|',expand=True)
```

```python
df1[['hero','publisher']]= df1['Hero|Publisher'].str.split('|',expand=True)
```

```python
df1.head(2)
```

```python
df1=df1.drop(columns=['Hero|Publisher'])
df1.head(2)
```

```python

to_replace = ['{','}']

for char in to_replace:
    df1['Measurements'] = df1['Measurements'].str.replace(char,'',regex=False)
    
df1['Measurements'].head()

```

```python
df1[['h1','w1']]= df1['Measurements'].str.split(',',expand=True)
df1=df1.drop(columns=['Measurements'])
df1.head(2)
```

```python
df1[['whaat','height(cm)','YAA']]= df1['h1'].str.split(' ',expand=True)
df1=df1.drop(['h1','YAA','whaat'],axis=1)
df1.head(2)
```

```python
df1[['w','woah','weight(kg)','w2']]= df1['w1'].str.split(' ',expand=True)
df1=df1.drop(['w','woah','w2','w1'],axis=1)
df1.head(2)
```

```python

to_replace = ["'"]

for char in to_replace:
    df1['height(cm)'] = df1['height(cm)'].str.replace(char,'',regex=False)
    df1['weight(kg)'] = df1['weight(kg)'].str.replace(char,'',regex=False)
    
df1.head()

```

### df1 cleaing complete

```python
df2.head()
```

```python
df2['Powers']=df2['Powers'].str.split(',')
```

```python
df2.head()
```

```python
df2['Powers']=df2['Powers'].astype(str)
```

```python
df2.loc [2,'Powers']
```

```python
df2['powersplit']=df2['Powers'].str.replace("'",'"')
```

```python
df2.loc[2,'powersplit']
```

```python
df2['powersplit']=df2['Powers'].str.replace("'",'"')
```

```python
df2['powersplit']=df2['powersplit'].apply(json.loads)
df2['powersplit'].head()
```

```python
df2['powersplit'].value_counts()
```

```python
exploded=df2.explode('powersplit')
exploded[['hero_names','Powers','powersplit']].head()
```

```python
df2.head()
```

```python
power_cols=exploded['powersplit'].dropna().unique()
power_cols
```

```python
for col in power_cols:
    df2[col]= df2['Powers'].str.contains(col)
df2.head()



```

```python
df2= df2.drop(columns=['Powers','powersplit'])
```

```python
df2.head()
```

### df2 clean/ohe complete

```python
df1.head()
```

```python
df2.head()
```

```python
df=pd.merge(df1,df2,left_on='hero',right_on='hero_names')
```

```python
df.head(3)
```

```python
df.to_csv('Data/hero_power.csv',index=False)
```
